# Simulation of the seeking PCA process.

```r
# Rotation rownames
my_dim = colnames(df)
my_dim
```

```r
# Rotation colnames 
my_PC = paste(rep('PC',ncol(df)), 1:ncol(df), sep = '')
my_PC
```

```r
X = t(data.matrix(df))
X
```

```r
center = rowMeans(X)
```

```r
# calculate the mean deviation form
B <- sweep(X,1,center)
B
```

```r
# according to the formula to find Covariance matrix
Sx <- (1/(ncol(B)-1))* B %*% t(B)
```

```r
# find eigenvalues and eigen vectors of the Sx
eigen(Sx)
```

```r
variances = eigen(Sx)$values; variances
```

```r
P = eigen(Sx)$vectors;P
```

```r
colnames(P) <- my_PC
rownames(P) <- my_dim
P
```

```r
V_proportion <- variances / sum(variances);V_proportion
```

```r
Cum_proportion <- cumsum(V_proportion); Cum_proportion
```

```r
thresh <- 0.98
thresh
```

```r
number <- length(V_proportion[Cum_proportion < thresh]) + 1; number
```

```r
PCA <- P[,1:number]
PCA
```

```r
transformed <- t(t(PCA) %*% B);transformed
```

```r
myPCA = function(df, thresh=0.9){
    
    my_dim <- colnames(df)
    my_PC = paste(rep('PC',ncol(df)), 1:ncol(df), sep = '')
    X = t(data.matrix(df))
    center = rowMeans(X)
    B <- sweep(X,1,center)
    Sx <- (1/(ncol(B)-1))* B %*% t(B)
    
    variances = eigen(Sx)$values
    
    P = eigen(Sx)$vectors
    colnames(P) <- my_PC
    rownames(P) <- my_dim
    
    V_proportion <- variances / sum(variances);V_proportion
    Cum_proportion <- cumsum(V_proportion)
    
    number <- length(V_proportion[Cum_proportion < thresh]) + 1
    PCA <- P[,1:number]
    
    transformed <- t(t(PCA) %*% B)
    
    Output <- list(number=number, PCA=PCA, transformed=transformed)
    return(Output)
}
```

```r
myPCA(df, thresh = 0.98)
```
